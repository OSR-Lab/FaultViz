## Import the faultviz package

```python
import faultviz
import random
import tqdm.notebook
import time
```

## Start the visualization service

- Requires a JDK to be installed with JAVA_HOME correctly set in the PATH environment variable

```python
faultviz.start_view_service() # If started multiple times, the port may already be in use; set the port parameter to change the port number
```

```python
vt = faultviz.ViewWidget()
```

```python
vt.show()
```

```python
def attack():
    delay = random.randint(100,2000)
    pulse = random.randint(10, 50)
    state = random.choice(['s1', 's3', 'ssss'])
    t = random.random()
    data = 'rreerse'
    vt.update(state = state, delay = delay, pulse = pulse, data = data, t = t)
```

```python
attack() # After one update call, vt can be viewed normally
```

```python
vt.show()
```

```python
for i in tqdm.notebook.trange(100):
    time.sleep(0.5)
    attack() # Run the attack multiple times
```

```python
vt.plot()
```
